```python
import requests
import pandas as pd
import openpyxl
from bs4 import BeautifulSoup
```

```python
#declare variables, url, search, location, params
url = "https://www.yellowpages.com/"

search = 'dentists'
location = 'New York, NY'

params = {'search_terms': search, 'geo_location_terms': location}

url
```

```python
#make request
response = requests.get(url + 'search', params=params)
```

```python
#check status code
response
```

```python
#check url
data = response.text

#data
```

```python
#soup is used to grab elements off a page
soup = BeautifulSoup(data, 'html.parser')
```

```python
#example to show what soup does

#for link in soup.find_all('a'):
#    print(link.get('href'))
```

```python
#this is the class that contains the name of the dentist
dentists = soup.find_all('div', class_ = 'info')

#dentists
```

```python
#initializing an empty list
dentists_data = []

#looping through the dentists list and grabbing the name, address, phone, and website
for dentist in dentists:
    name =  dentist.find('a', class_='business-name').text
    #print(name)

    address_element = dentist.find('p', class_='adr') or dentist.find('div', class_='adr')
    address = address_element.text if address_element else 'No Address'
    #print(address)

    phone_element = dentist.find('div', class_='phone') or dentist.find('div', class_='phones phone primary')
    phone = phone_element.text if phone_element else 'No Phone'
    #print(phone)

    website_element = dentist.find('div', class_='links').find('a', text='Website')
    website = website_element['href'] if website_element else 'No Website'
    #print(website)

    #creating a dictionary
    dentist_info = {
        'Name': name,
        'Address': address,
        'Phone': phone,
        'Website': website,
    }
    
    #appending the dictionary to the list
    dentists_data.append(dentist_info)
```

```python
#displaying the list
dentists_data
```

```python
#Create a Pandas df
dentist_df = pd.DataFrame(dentists_data)
```

```python
#display the df
dentist_df
```

```python
#export the df to excel
dentist_df.to_excel('dentist_info.xlsx', index=False)
```
